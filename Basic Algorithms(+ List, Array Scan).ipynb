### 배열 원소를 역순으로 정렬

```python
from typing import Any, MutableSequence

def reverse_array(a : MutableSequence) -> None:
    
    n = len(a)
    for i in range(n // 2):
        a[i], a[n - i - 1] = a[n - i - 1], a[i] # 각 원소를 서로 바꿔줌 
        
if __name__ == '__main__': # 이게 맞지 않으면 실행 x
    print('배열 원소를 역순으로 정렬')
    nx = int(input("원소 수 입력"))
    x = [None] * nx 
    # 원소 수가 NX인 만큼의 리스트 생성
    
    for i in range(nx):
        x[i] = int(input(f'x[{i}]값을 입력하세요 : ')) # nx의 개수만큼 직접 원소 입력 
        
    reverse_array(x)
    # 받은 원소를 역순으로 정렬 
    
    for i in range(nx):
        print(f'x[{i}] = {x[i]}')
        
# 원소를 역순으로 정렬할 수 있게 되었다.
```

### 기수 변환(N 진수구하기) 
    - 정수값을 임의의 기수로 변환하는 알고리즘 
    - 기수 : 수를 나타내는 기초이며 10진법은 0 ~ 9까지의 정수
    - 서수 : 사물의 순서를 나타냄 

```python
def card_conv(x : int, r : int) -> str: # 숫자를 문자로 반환 
    
    d = ''
    dchar = '0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ'
    
    while x > 0:
        d += dchar[x % r] # x를 r로 나눈 나머지를 인덱스로 
        x //= r
        # x가 0이 될 때까지 진행  
        
    return d[::-1]
```

```python
if __name__ == '__main__':
    print('10진수를 n진수로 변환')
    
    while True:
        while True:
            no = int(input('변환할 값으로 음이 아닌 정수 입력 : '))
            if no > 0:
                break
                # 양수가 들어오면 위의 과정을 중단 
                
        while True:
            cd = int(input('어떤 진수로 변환?'))
            if 2 <= cd <= 36 :
                break
                # 2 ~ 36진수까지 입력이 되면 중단 
                
        print(f'{cd}진수로는 {card_conv(no, cd)}입니당.')
        
        retry = input('One more? (Y / N)')
        if retry in {'N', 'n'}:
            break
```

```python
def sum_1(n):
    
    s = 0
    while n > 0:
        s += n
        n -= 1
    return s

x = int(input('x값 입력'))
print(f'1부터 {x}까지 정수 합은 {sum_1(x)}입니다.')

# 매개변수 n은 1씩 감소하는데 x는 감소하지 않는다.
# 즉 매개변수에 실제 인수가 대입되는 것 <- 중요 
```

### 소수 나열
    - 어떤 정수 이하의 소수를 모두 나열하는 알고리즘 
    * 소수 : 1과 자신 이외의 정수로 나누어 떨어지지 않는 정수 

```python
counter = 0

for n in range(2, 1001):
    for i in range(2, n):
        counter += 1
        if n % i == 0: # 나누어 떨어지면 소수가 아님 자신 이외에는 나누어 떨어지면 안되기 때문 
            break
            # 나누어 떨어지면 나누기 중단 
    else:
        print(n)
        # 소수만 출력 
        
print(f'나눗셈을 실행한 횟수 : {counter}')
```

```python
# 1번째 개선 알고리즘 

counter = 0
ptr = 0
prime = [None] * 500

prime[ptr] = 2 # 초기값으로 소수인 2를 지정 
ptr += 1

for n in range(3, 1001, 2):
    for i in range(1, ptr):
        counter += 1
        if n % prime[i] == 0: 
            # 저장된 소수들과 나눠지면 소수가 아니기 때문에 중단되고 저장된 소수와 나눠지지 않으면 소수이므로 prime에 인덱스와 값을 저장 
            break
    else:
        prime[ptr] = n
        ptr += 1
        
for i in range(ptr):
    print(prime[i])
    
print(f'나눗셈을 실행한 횟수 : {counter}')

```

```python
# 2번째 개선 알고리즘 (n의 제곱근 이하의 어떤 소수로도 나누어 떨어지지 않으면 그것은 소수다 라느 아이디어)

counter = 0
ptr = 0
prime = [None] * 500

prime[ptr] = 2
ptr += 1

prime[ptr] = 3
ptr += 1

for n in range(5, 1001, 2):
    i = 1
    while prime[i] * prime[i] <= n:
        counter += 2
        if n % prime[i] == 0:
            break
        i+= 1
    else:
        prime[ptr] = n
        ptr += 1
        counter += 1
        
for i in range(ptr):
    print(prime[i])
    
print(f'곱셈과 나눗셈을 실행한 횟수 : {counter}')

# 실행 횟수가 눈에 띄게 줄어든 것을 볼 수 있다.
```
