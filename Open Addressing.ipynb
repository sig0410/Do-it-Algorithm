```python
from enum import Enum
from open_hash import OpenHash

Menu = Enum('Menu', ['', '', '', '', ''])

def select_menu() -> Menu:
    
    s = [f'({m.value}){m.name}' for m in Menu]
    while True:
        print(*s, sep = '  ', end = '')
        n = int(input(': '))
        if 1 <= n <= len(Menu):
            return Menu(n)
        
hash = OpenHash(13)

while True:
    menu = select_menu()
    
    if menu == Menu.:
        key = int(input('   : '))
        val = input('   : ')
        if not hash.add(key, val):
            print('  .')
            
    elif menu == Menu.:
        key = int(input('   : '))
        if not hash.remove(key):
            print(' .')
            
    elif menu == Menu.:
        key = int(input('   : '))
        t = hash.search(key)
        if t is not None:
            print(f'    {t}')
        else:
            print('  .')
            
    elif menu == Menu.:
        hash.dump()
        
    else:
        break
```
